
**step 1- indexing (document augmentation)**

```python
video_id="n-Xkbqcfi9w"
try:
  transcript_list=YouTubeTranscriptApi.get_transcript(video_id,languages=['en'])
  transcript=" ".join([chunk['text'] for chunk in transcript_list])
  print(transcript)
except TranscriptsDisabled:
  print(" no caption available for this video")
```

```python
transcript_list
```

**step 1.b use text splitter to break int o small semantic chunks**

```python
splitter=RecursiveCharacterTextSplitter(chunk_size=250,chunk_overlap=30)
```

```python
chunks=splitter.create_documents([transcript])
```

```python
#we have used create_documents so that we get output in documnet format
```

```python
len(chunks)
```

```python
chunks[0]
```

**stpe 1.c&1.d - convert these chunks to vector and store them to vector store**

```python
embeddings=GoogleGenerativeAIEmbeddings(model="models/embedding-001",google_api_key=os.environ["GEMINI_API_KEY"])
```

```python
vector_store=FAISS.from_documents(chunks,embeddings)
```

```python
vector_store.index_to_docstore_id
```

**step 2- retrieval**

```python
retriever=vector_store.as_retriever(search_kwargs={"k":4},search_type="similarity")
```

```python
retriever
```

```python
retriever.invoke("what is competitive programming")
```

**step 3 - augmentation**

```python
llm=ChatGoogleGenerativeAI(model="models/gemini-1.5-flash",google_api_key=os.environ["GEMINI_API_KEY"])
```

```python
prompt=PromptTemplate(
    template="""
      you are a helpful assistant.
      answer only from the provided transcript context.
      if the context is insufficient ,just say you don't know.
      {context}
      Question: {Question}
      """,
    input_variables=["context","Question"]
)
```

```python
Question="is the topic abiut dsa covered in this video? if yes than what was discussed"
retrieved_document=retriever.invoke(Question)
```

```python
context_text="\n\n".join(doc.page_content for doc in retrieved_document)
```

```python
context_text
```

```python
final_prompt=prompt.invoke({"context":context_text,"Question":Question})
```

**step-4 :generation**

```python
answer=llm.invoke(final_prompt)
print(answer.content)
```

**building a chain**

```python
from langchain_core.runnables import RunnableLambda,RunnableParallel,RunnablePassthrough
```

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
def format_docs(retrieved_docs):
  context_text="\n\n".join(doc.page_content for doc in retrieved_docs)
  return context_text
```

```python
parallel_chain=RunnableParallel({
    'context':retriever|RunnableLambda(format_docs),
    'Question':RunnablePassthrough()

})
```

```python
parallel_chain.invoke("tell me about competitive programming")
```

```python
parser=StrOutputParser()
```

```python
main_chain=parallel_chain|prompt|llm|parser
```

```python
main_chain.invoke("can you summarize this video")
```
